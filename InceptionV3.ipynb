```python
#building the InceptionV3 model
from tensorflow import keras
from keras.layers import Input
from keras import backend as K
from keras import layers
from keras import layers
from keras.layers import *
from keras.models import Model
from tensorflow.keras.layers import Input

def cnn_conv2d(x,
              filters,
              rows,
              columns,
              padding='same',
              strides=(1, 1),
              name=None):

    x = Conv2D(
        filters, (rows, columns),
        strides=strides,
        padding=padding,
        use_bias=False,
        name=None)(x)
    x = BatchNormalization(axis=3, scale=False, name=None)(x)
    x = Activation('relu', name=name)(x)
    return x


def InceptionV3(include_top=True,
                input_tensor=None,
                input_shape=None,
                pooling=None,
                classes=26):
    inputs = Input(shape=(299,299,3)) 

    x = cnn_conv2d(inputs, 32, 3, 3, strides=(2, 2), padding='valid')
    x = cnn_conv2d(x, 32, 3, 3, padding='valid')
    x = cnn_conv2d(x, 64, 3, 3)
    x = MaxPooling2D((3, 3), strides=(2, 2))(x)

    x = cnn_conv2d(x, 80, 1, 1, padding='valid')
    x = cnn_conv2d(x, 192, 3, 3, padding='valid')
    x = MaxPooling2D((3, 3), strides=(2, 2))(x)

    #35 x 35 x 256
    branchLayer1x1 = cnn_conv2d(x, 64, 1, 1)

    branchLayer5x5 = cnn_conv2d(x, 48, 1, 1)
    branchLayer5x5 = cnn_conv2d(branchLayer5x5, 64, 5, 5)

    branchLayer3x3_1 = cnn_conv2d(x, 64, 1, 1)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)

    branch_pool_Layer = AveragePooling2D((3, 3), strides=(1, 1), padding='same')(x)
    branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 32, 1, 1)
    x = layers.concatenate(
        [branchLayer1x1, branchLayer5x5, branchLayer3x3_1, branch_pool_Layer],
        axis=3,
        name='layer0')

    #35 x 35 x 256
    branchLayer1x1 = cnn_conv2d(x, 64, 1, 1)

    branchLayer5x5 = cnn_conv2d(x, 48, 1, 1)
    branchLayer5x5 = cnn_conv2d(branchLayer5x5, 64, 5, 5)

    branchLayer3x3_1 = cnn_conv2d(x, 64, 1, 1)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)

    branch_pool_Layer = AveragePooling2D((3, 3), strides=(1, 1), padding='same')(x)
    branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 64, 1, 1)
    x = layers.concatenate(
        [branchLayer1x1, branchLayer5x5, branchLayer3x3_1, branch_pool_Layer],
        axis=3,
        name='layer1')

    #35 x 35 x 256
    branchLayer1x1 = cnn_conv2d(x, 64, 1, 1)

    branchLayer5x5 = cnn_conv2d(x, 48, 1, 1)
    branchLayer5x5 = cnn_conv2d(branchLayer5x5, 64, 5, 5)

    branchLayer3x3_1 = cnn_conv2d(x, 64, 1, 1)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)

    branch_pool_Layer = AveragePooling2D((3, 3), strides=(1, 1), padding='same')(x)
    branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 64, 1, 1)
    x = layers.concatenate(
        [branchLayer1x1, branchLayer5x5, branchLayer3x3_1, branch_pool_Layer],
        axis=3,
        name='layer2')

    #17 x 17 x 768
    branchLayer3x3 = cnn_conv2d(x, 384, 3, 3, strides=(2, 2), padding='valid')

    branchLayer3x3_1 = cnn_conv2d(x, 64, 1, 1)
    branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 96, 3, 3)
    branchLayer3x3_1 = cnn_conv2d(
        branchLayer3x3_1, 96, 3, 3, strides=(2, 2), padding='valid')

    branch_pool_Layer = MaxPooling2D((3, 3), strides=(2, 2))(x)
    x = layers.concatenate(
        [branchLayer3x3, branchLayer3x3_1, branch_pool_Layer], axis=3, name='layer3')

    #17 x 17 x 768
    branchLayer1x1 = cnn_conv2d(x, 192, 1, 1)

    branchLayer7x7 = cnn_conv2d(x, 128, 1, 1)
    branchLayer7x7 = cnn_conv2d(branchLayer7x7, 128, 1, 7)
    branchLayer7x7 = cnn_conv2d(branchLayer7x7, 192, 7, 1)

    branchLayer7x7_1 = cnn_conv2d(x, 128, 1, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 128, 7, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 128, 1, 7)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 128, 7, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 1, 7)

    branch_pool_Layer = AveragePooling2D((3, 3), strides=(1, 1), padding='same')(x)
    branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 192, 1, 1)
    x = layers.concatenate(
        [branchLayer1x1, branchLayer7x7, branchLayer7x7_1, branch_pool_Layer],
        axis=3,
        name='layer4')

    #17 x 17 x 768
    for i in range(2):
        branchLayer1x1 = cnn_conv2d(x, 192, 1, 1)

        branchLayer7x7 = cnn_conv2d(x, 160, 1, 1)
        branchLayer7x7 = cnn_conv2d(branchLayer7x7, 160, 1, 7)
        branchLayer7x7 = cnn_conv2d(branchLayer7x7, 192, 7, 1)

        branchLayer7x7_1 = cnn_conv2d(x, 160, 1, 1)
        branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 160, 7, 1)
        branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 160, 1, 7)
        branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 160, 7, 1)
        branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 1, 7)

        branch_pool_Layer = AveragePooling2D(
            (3, 3), strides=(1, 1), padding='same')(x)
        branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 192, 1, 1)
        x = layers.concatenate(
            [branchLayer1x1, branchLayer7x7, branchLayer7x7_1, branch_pool_Layer],
            axis=3,
            name='layer' + str(5 + i))

    #17 x 17 x 768
    branchLayer1x1 = cnn_conv2d(x, 192, 1, 1)

    branchLayer7x7 = cnn_conv2d(x, 192, 1, 1)
    branchLayer7x7 = cnn_conv2d(branchLayer7x7, 192, 1, 7)
    branchLayer7x7 = cnn_conv2d(branchLayer7x7, 192, 7, 1)

    branchLayer7x7_1 = cnn_conv2d(x, 192, 1, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 7, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 1, 7)
    branch7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 7, 1)
    branchLayer7x7_1 = cnn_conv2d(branchLayer7x7_1, 192, 1, 7)

    branch_pool_Layer = AveragePooling2D((3, 3), strides=(1, 1), padding='same')(x)
    branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 192, 1, 1)
    x = layers.concatenate(
        [branchLayer1x1, branchLayer7x7, branchLayer7x7_1, branch_pool_Layer],
        axis=3,
        name='layer7')

    #8 x 8 x 1280
    branchLayer3x3 = cnn_conv2d(x, 192, 1, 1)
    branchLayer3x3 = cnn_conv2d(branchLayer3x3, 320, 3, 3,
                          strides=(2, 2), padding='valid')

    branchLayer7x7x3 = cnn_conv2d(x, 192, 1, 1)
    branchLayer7x7x3 = cnn_conv2d(branchLayer7x7x3, 192, 1, 7)
    branchLayer7x7x3 = cnn_conv2d(branchLayer7x7x3, 192, 7, 1)
    branchLayer7x7x3 = cnn_conv2d(
        branchLayer7x7x3, 192, 3, 3, strides=(2, 2), padding='valid')

    branch_pool_Layer = MaxPooling2D((3, 3), strides=(2, 2))(x)
    x = layers.concatenate(
        [branchLayer3x3, branchLayer7x7x3, branch_pool_Layer], axis=3, name='layer8')

    #8 x 8 x 2048
    for i in range(2):
        branchLayer1x1 = cnn_conv2d(x, 320, 1, 1)

        branchLayer3x3 = cnn_conv2d(x, 384, 1, 1)
        branchLayer3x3_1 = cnn_conv2d(branchLayer3x3, 384, 1, 3)
        branchLayer3x3_2 = cnn_conv2d(branchLayer3x3, 384, 3, 1)
        branchLayer3x3 = layers.concatenate(
            [branchLayer3x3_1, branchLayer3x3_2], axis=3, name='layer9' + str(i))

        branchLayer3x3 = cnn_conv2d(x, 448, 1, 1)
        branchLayer3x3_1 = cnn_conv2d(branchLayer3x3_1, 384, 3, 3)
        branchLayer3x3_11 = cnn_conv2d(branchLayer3x3_1, 384, 1, 3)
        branchLayer3x3_22 = cnn_conv2d(branchLayer3x3_1, 384, 3, 1)
        branchLayerLayer3x3_1 = layers.concatenate(
            [branchLayer3x3_11, branchLayer3x3_22], axis=3)

        branch_pool_Layer = AveragePooling2D(
            (3, 3), strides=(1, 1), padding='same')(x)
        branch_pool_Layer = cnn_conv2d(branch_pool_Layer, 192, 1, 1)
        x = layers.concatenate(
            [branchLayer1x1, branchLayer3x3, branchLayer3x3_1, branch_pool_Layer],
            axis=3,
            name='layer' + str(9 + i))
    if include_top:
        # classification
        x = GlobalAveragePooling2D(name='avg_pool')(x)
        x = Dense(classes, activation='softmax', name='predictions')(x)
    else:
        if pooling == 'avg':
            x = GlobalAveragePooling2D()(x)
        elif pooling == 'max':
            x = GlobalMaxPooling2D()(x)


    model = Model(inputs, x, name='inception_v3')
    return model,x,inputs
 
```

```python
#training the model
from re import X
from keras.layers import Input
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation,BatchNormalization
from tensorflow.keras.layers import Flatten, Dropout
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import backend as K
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
from keras.preprocessing import image


ImgH, ImgW = 299, 299
batch_size = 64
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='models/CNN_inception/model.h5',
    monitor='val_acc',
    save_freq='epoch')


train_datagen = ImageDataGenerator(
		rescale=1./255)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
	directory=r"./final_split1/train/",
	target_size=(ImgH, ImgW),
	color_mode="rgb",
	batch_size=batch_size,
	class_mode="categorical"
)


test_generator = test_datagen.flow_from_directory(
	directory=r"./final_split1/test/",
	target_size=(ImgH, ImgW),
	color_mode="rgb",
	batch_size=batch_size,
	class_mode="categorical"
)

classes_list = train_generator.class_indices
print(classes_list,len(classes_list))
N = len(classes_list)

f = open('modelSGD0.1/CNN_inception/LABELS.txt','a+')
lst = list(classes_list.keys())
f.write('\n'.join(lst))
f.close()


model,x,inputs = InceptionV3()
print(model.summary())

model = load_model('modelSGD0.1/CNN_inception/model.h5')
print('Model Loaded!!')
opt =  tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss='categorical_crossentropy', optimizer=opt,metrics=['accuracy'])

hist=model.fit_generator(
        train_generator,
        epochs=5,
        validation_data=test_generator,verbose=1,
        callbacks=[model_checkpoint_callback])
model.save_weights('modelSGD0.1/CNN_inception/first_try.h5')

#plots the graph
plt.plot(hist.history['val_accuracy'])
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","Validation Accuracy","loss","Validation Loss"])
plt.show(block=True)
```

```python
#testing data
import cv2
import os
from keras.models import load_model
import numpy as np
dim = (299,299)
model = load_model('model4/CNN_inception/model.h5')
LABELS = open('model4/CNN_inception/LABELS.txt').read().split('\n')
print(model.summary())
path  = 'final_split1/test'
y_pred=[]
y_act=[]
for gt in os.listdir(path):
  matrix = {'TP':0, 'FP':0, 'TN': 0, 'FN':0}
  for imgf in os.listdir(os.path.join(path,gt)):
    img = cv2.imread(os.path.join(path,gt,imgf))
    im_rgb = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    im_rgb = cv2.resize(im_rgb, dim, interpolation = cv2.INTER_AREA)
    im_rgb = np.array(im_rgb)/255
    im = np.expand_dims(im_rgb, axis=0)
    res = model.predict(im)
    pred = LABELS[np.argmax(res[0])]
    y_pred.append(pred)
    y_act.append(gt)
    if pred == gt:
      matrix['TP'] +=1
    else:
      matrix['FP'] +=1
  print(gt,matrix['TP'])
  #acurracy 
  if matrix['TP']!=0:
    Precision=matrix['TP']/(matrix['TP']+matrix['FP'])
    # Precision
    Recall=matrix['TP']/(matrix['TP']+matrix['FN'])
    #Recall
    print('Accuracy: ',matrix['TP']/len(os.listdir(os.path.join(path,gt))),'Precision: ',Precision,'Recall: ',Recall,'F1-Score: ',2*(Recall * Precision) / (Recall + Precision))
  else:
    print('Accuracy: ',matrix['TP']/len(os.listdir(os.path.join(path,gt))))
```

```python
#overall accoracy
from sklearn.metrics import accuracy_score
accuracy_score(y_act, y_pred)
```
